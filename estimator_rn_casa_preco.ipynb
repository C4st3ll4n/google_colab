```
colunas_usadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```
base = pd.read_csv('house-prices.csv', usecols=colunas_usadas)
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
scalerX = MinMaxScaler()
```

```
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scalerX.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```
base.head()
```

```
scalerY = MinMaxScaler()
```

```
base[['price']] = scalerY.fit_transform(base[['price']])
base.head()
```

```
X = base.drop('price', axis=1)
Y = base.price
```

```
previsores_coluna = colunas_usadas[1:17]
previsores_coluna
```

```
import tensorflow as tf
```

```
colunas = [tf.feature_column.numeric_column(c) for c in previsores_coluna]
```

```
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size=0.2)
```

```
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(X, Y, num_epochs=None, shuffle=True)
```

```
regressor = tf.estimator.DNNRegressor([8, 8, 8, 8],colunas)
regressor.train(funcao_treinamento, steps=30000)
```

```
funcao_prev = tf.estimator.inputs.pandas_input_fn(X, shuffle = False)
previssoes = regressor.predict(funcao_prev)
```

```
list(previssoes)
```

```
previsoes = []
```

```
for p in regressor.predict(funcao_prev):
    previsoes.append(p['predictions'][0])
```

```
import numpy as np
```

```
val_prev = np.array(previsoes).reshape(-1,1)
```

```
val_prev = scalerY.inverse_transform(val_prev)
```

```
Y_teste2 = scalerY.inverse_transform(Y_teste.values.reshape(-1,1))
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error
```
